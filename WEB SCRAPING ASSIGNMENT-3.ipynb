4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on 
www.flipkart.com and scrape following details for all the search results displayed on 1st page. 
Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, 
“Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display 
Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”. 
Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe 
and CSV.


```python
# Lets first connect to web driver
driver = webdriver.Chrome(r"C:\Users\ASUS\Downloads\chromedriver_win32\chromedriver.exe")
time.sleep(4)

url = 'https://www.flipkart.com/'
driver.get(url)
time.sleep(6)

#Searching for a samrtphone by giving an input in the search box
search = driver.find_element_by_xpath("//input[@class='_3704LK']")
search.clear()
search.send_keys("iphone 12 pro")

#Pressing the search button for the results
search_btn = driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
search_btn.click()
```

```python
#list of all the specifications in the mobile
Title=[]
RAM=[]
Camera=[]
Display_Size=[]
Processor=[]
Battery_Capacity=[]
Price=[]
Product_URL=[]
```

```python
# Scraping all the product names
title=driver.find_elements_by_xpath('//div[@class="_4rR01T"]')
for i in title:
    if i.text is None :
        Title.append("--") 
    else:
        Title.append(i.text)
        
# Scraping all the product RAM       
Ram=driver.find_elements_by_xpath("//div[@class='fMghEO']/ul/li[1]")
for i in Ram:
    if i.text is None :
        RAM.append("--") 
    else:
        RAM.append(i.text)
        
# Scraping all the product Cameras(i.e., Primary and Secondary)
cam=driver.find_elements_by_xpath("//div[@class='fMghEO']/ul/li[3]")
for i in cam:
    if i.text is None :
        Camera.append("--") 
    else:
        Camera.append(i.text)

# Scraping all the products display
display=driver.find_elements_by_xpath("//div[@class='fMghEO']/ul/li[2]")
for i in display:
    if i.text is None :
        Display_Size.append("--") 
    else:
        Display_Size.append(i.text)

# Scraping all the products processor
processor=driver.find_elements_by_xpath("//div[@class='fMghEO']/ul/li[5]")
for i in processor:
    if i.text is None :
        Processor.append("--") 
    else:
        Processor.append(i.text)
        
# Scraping all the products Battery
Battery=driver.find_elements_by_xpath("//div[@class='fMghEO']/ul/li[4]")
for i in Battery:
    if i.text is None :
        Battery_Capacity.append("--") 
    else:
        Battery_Capacity.append(i.text)

# Scraping all the products price
price = driver.find_elements_by_xpath("//div[@class='_30jeq3 _1_WHN1']")
for i in price:
    if i.text is None :
        Price.append("--") 
    else:
        Price.append(i.text)

# Scraping all the product urls
url="https://www.flipkart.com/"
url1=driver.find_element_by_xpath("//a[@class='_1fQZEK']").get_attribute('href')
for i in url:
    #URL.append(i.text)
    Product_URL.append(url1)
```

6. Write a program to scrap details of all the funding deals for second quarter (i.e. July 20 –
September 20) from trak.in.

```python
# Lets first connect to web driver
driver = webdriver.Chrome(r"C:\Users\ASUS\Downloads\chromedriver_win32\chromedriver.exe")
time.sleep(4)

url = 'hhttps://trak.in/india-startup-funding-investment-2015/'
driver.get(url)
time.sleep(6)
```

```python
fund = {}
fund['Date'] = []
fund['Startup Name'] = []
fund['Industry/Vertical'] = []
fund['Sub-Vertical'] = []
fund['City/Location'] = []
fund['Investor Name'] = []
fund['Investment Type'] = []
fund['Amount(in USD)'] = []
```

```python
for i in range(48,51):
    driver.find_element_by_xpath('//div[@id="tablepress-{}_wrapper"]/div/label/select/option[4]'.format(i)).click()

    # For Date
    dt = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[2]'.format(i))
    for d in dt:
        fund['Date'].append(d.text)
    # For Startup Name
    sn = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[3]'.format(i))
    for n in sn:
        fund['Startup Name'].append(n.text)
    # For Industry/Vertical
    ind = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[4]'.format(i))
    for n in ind:
        fund['Industry/Vertical'].append(n.text)
    # For Sub-Vertical
    sv = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[5]'.format(i))
    for s in sv:
        fund['Sub-Vertical'].append(s.text)
    # For City/Location
    loc = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[6]'.format(i))
    for l in loc:
        fund['City/Location'].append(l.text)
     # For Investor Name
    inv = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[7]'.format(i))
    for n in inv:
        fund['Investor Name'].append(n.text)
     # Investment Type
    invt = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[8]'.format(i))
    for n in invt:
        fund['Investment Type'].append(n.text)
     # Amount(in USD)
    amt = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[9]'.format(i))
    for a in amt:
        fund['Amount(in USD)'].append(a.text)
    
fund_df = pd.DataFrame(fund)
fund_df
```

 Write a program to scrap all the available details of top 10 gaming laptops from digit.in.

```python
# Lets first connect to web driver
driver = webdriver.Chrome(r"C:\Users\ASUS\Downloads\chromedriver_win32\chromedriver.exe")
time.sleep(4)

url = 'https://www.digit.in/'
driver.get(url)
time.sleep(6)
```

```python
# Opening the top ten laptops
topten = driver.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[1]/div[5]/div[1]/div[4]/div[2]/a/p")
topten.click()
```

```python
name = []
price = []
OS = []
display = []
processor = []
HDD = []
RAM = []
weight = []
dimension = []
GPU = []
```

```python
# names
names=driver.find_elements_by_xpath("//div[@class='right-container']/div/a/h3")
for i in names:
    name.append(i.text)    
# os
os=driver.find_elements_by_xpath("//div[@class='product-detail']/div/ul/li[1]/div/div")
for i in os:
    OS.append(i.text)    
# display
displays=driver.find_elements_by_xpath("//div[@class='product-detail']/div/ul/li[2]/div/div")
for i in displays:
    display.append(i.text)    
# processor
processors=driver.find_elements_by_xpath("//div[@class='product-detail']/div/ul/li[3]/div/div")
for i in processors:
    processor.append(i.text)
# memory
memories=driver.find_elements_by_xpath("//div[@class='Spcs-details'][1]/table/tbody/tr[6]/td[1]")
memories_spec=driver.find_elements_by_xpath("//div[@class='Spcs-details'][1]/table/tbody/tr[6]/td[3]")
for i in range(len(memories)):
        if memories[i].text=='Memory':
            HDD.append(memories_spec[i].text.split('/')[0])
            RAM.append(memories_spec[i].text.split('/')[1])
        else:
            HDD.append('No details')
            RAM.append('No details')
# weight
weights=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr/td[1]")
weight_spec=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr/td[3]")
for i in range(len(weights)):
        if weights[i].text=='Weight':
            weight.append(weight_spec[i].text)        
# dimension
dimension=[]
dims=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr/td[1]")
dims_spec=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr/td[3]")
for i in range(len(dims)):
        if dims[i].text=='Dimension':
            dimension.append(dims_spec[i].text)
# graphical processor
GPUs=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr/td[1]")
GPUs_spec=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr/td[3]")
for i in range(len(GPUs)):
        if GPUs[i].text=='Graphics Processor':
            GPU.append(GPUs_spec[i].text)     
```

```python
#getting the url of full specs links
full_specs=[]
urls=driver.find_elements_by_xpath("//div[@class='full-specs']/span")
for i in urls:
    if i.get_attribute('data-href'):
        full_specs.append(i.get_attribute('data-href'))
#iterating throug every laptops full specs' page    
for i in full_specs:
    driver.get(i)
    try:
        prices=driver.find_element_by_xpath("//div[@class='Block-price']/b")
        price.append(prices.text)
    except NoSuchElementException:
        price.append("No details")
```

```python
laptop=pd.DataFrame({"Name":name,
                "Price":price,
                "OS":OS,
                "Display":display,
                "HDD":HDD,
                 "RAM":RAM,
                "processor":processor,
                "weight":weight,
                "Dimension":dimension,
                "Graphical processor":GPU})
laptop
```
